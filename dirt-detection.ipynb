```python
from ultralytics import YOLO
dataset_path = "/kaggle/input/dirt-data/data.yaml" 
model = YOLO("yolov8m-seg.pt")
model.train(
    data=dataset_path,
    epochs=100,
    batch=16,
    imgsz=640,
    project="/kaggle/working/dirt_model",
    name="train1",
    device="cuda"
)
```

```python
from ultralytics import YOLO

model = YOLO("/kaggle/working/dirt_model/train1/weights/best.pt")

results = model.val(split="test")
```

Inferences on the images to check out which images in the test part are misclassified

```python
from ultralytics import YOLO

model = YOLO('/kaggle/working/dirt_model/train1/weights/best.pt')

test_images_path = '/kaggle/input/dirt-data/test/images'

results = model.predict(source=test_images_path, save=True, save_txt=True)

for r in results:
    print(r)  
```
